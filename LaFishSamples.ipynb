```python
import pandas as pd
from pandas import DataFrame
import numpy as np
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.cross_validation import train_test_split
from sklearn.cluster import KMeans
```

```python
fish_df = pd.read_excel('D:/Datasets/SCIO/FishSpecies/LA_fish1.xlsx', header = 0)
fish2_df = pd.read_excel('D:/Datasets/SCIO/FishSpecies/LA_fish2.xlsx', header = 0)
fish2_df
```

```python
fish_no_co = fish_df.loc[fish_df['co_added'] == 'COnotadded']
fish_no_co = fish_no_co.drop(['blood_side', 'co_added', 'Packaging', 'sampling_time', 'sample_id', 'id'], axis = 1)
```

```python
fish_temp = fish2_df.loc[fish2_df['Species'] == 'drum,black']
fish_temp2 = fish2_df.loc[fish2_df['Species'] == 'catfish,blue']
fish_temp3 = fish_no_co.loc[fish_no_co['species'] == 'Drum,Black']
fish_temp4 = fish_no_co.loc[fish_no_co['species'] == 'Catfish,Blue']

fish_temp.rename(columns = {'Species': 'species'}, inplace = True)
fish_temp2.rename(columns = {'Species': 'species'}, inplace = True)
```

```python
#fish_temp3['species'] = fish_temp3['species'].apply(lambda x: x.lower())
fishlist = [fish_temp, fish_temp2, fish_temp3, fish_temp4]

for l in fishlist:
    l.reset_index(inplace= True)
```

```python
fish_final = pd.concat([fish_temp, fish_temp2, fish_temp3, fish_temp4], axis = 0)

fish_final = fish_final.drop('index', axis = 1)
fish_final
```

```python
def species(x):
    if x == 'drum,black':
        x = 1
    elif x == 'drum,red':
        x = 2
    elif x == 'catfish,blue':
        x = 3
    return x

fish_label = fish_final['species'].apply(lambda x: species(x))
X = fish_final.ix[:, 1:]
#co_label = fish_df['co_added']
```

```python
fish_label.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, fish_label, test_size = .25, random_state = 43)
```

```python
grad_boost = GradientBoostingClassifier(n_estimators = 1000, max_depth = 4)
grad_boost.fit(X_train, y_train)
```

```python
grad_boost.score(X_test, y_test)
```

```python
X_pred = grad_boost.predict(X_test)
grad_pred = DataFrame(X_pred, y_test)
grad_pred.reset_index(inplace= True)
grad_pred.columns = ['Species', 'Predicted Species']
grad_pred
```

```python
r_forest = RandomForestClassifier()
r_forest.fit(X_train, y_train)
r_forest.score(X_test, y_test)
```
