```python
import os
import json
import sys
import pandas as pd
import numpy as np
from datetime import datetime
from sqlalchemy import create_engine
from pandas.io import sql
from glob import glob
from tqdm import tqdm

engine = create_engine('mysql+pymysql://root:v1933@127.0.0.1:3306/pga?charset=utf8')
conn = engine.connect()
```

```python
def turn_to_float(x):
    try:
        return(float(x))
    except:
        for letter in x[1:]:
            if letter != '.':
                try:
                    int(letter)
                except:
                    x = x.replace(letter, '')
                    
        
    return float(x)

os.chdir('/home/valesco/Datasets/PGA_Data/PGA_Dir/')

tournament_ls = glob('*')

player_holes_failed = ['013_2017']

for tourny in tqdm(player_holes_failed):
    
    #Player holes
    player_holes_df = pd.read_csv('/home/valesco/Datasets/PGA_Data/PGA_Dir/{}/{}_player_holes.csv'.format(tourny,tourny))
    player_holes_year = player_holes_df['year'].loc[0]
    player_holes_tourny_num = player_holes_df['tournament_id'].loc[0]

    player_holes_sql = 'SELECT * FROM pga.player_holes WHERE tournament_id = {} and year = {}'.format(player_holes_tourny_num,
                                                                                            player_holes_year)

    temp_df = pd.read_sql(player_holes_sql, engine)
    
    if len(temp_df) == 0:
        for i in range(1,12):
            col = 'z_coord_{}'.format(i)
            player_holes_df[col] = player_holes_df[col].apply(lambda x: turn_to_float(x))
            
        player_holes_df.to_sql('player_holes', engine, if_exists = 'append', index = False)
        
    #Course Holes
    course_holes_df = pd.read_csv('/home/valesco/Datasets/PGA_Data/PGA_Dir/{}/{}_course_holes.csv'.format(tourny,tourny))
    course_holes_year = player_holes_df['year'].loc[0]
    course_holes_tourny_num = player_holes_df['tournament_id'].loc[0]

    course_holes_sql = 'SELECT * FROM pga.course_holes WHERE tournament_id = {} and year = {}'.format(player_holes_tourny_num,
                                                                                            player_holes_year)
    
    course_temp_df = pd.read_sql(course_holes_sql, engine)
    
    if len(course_temp_df) == 0:
        course_holes_df.to_sql('course_holes', engine, if_exists = 'append', index = False)
```
