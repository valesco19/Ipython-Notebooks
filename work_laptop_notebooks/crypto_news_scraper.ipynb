```python
import pandas as pd
import time
from datetime import datetime, timedelta
import feedparser
import os
import requests
import json
```

```python
os.chdir('D:/data/crypto/news')

feeds = {
    'coindesk': 'http://feeds.feedburner.com/Coindesk?format=xml',
    'cointelegraph': 'https://cointelegraph.com/feed', 
    'CryptoCoinsNews': 'https://cryptocoinsnews.com/feed',
    'Blockchain-Blog': 'https://blog.blockchain.com/feed/',
    'Reddit Bitcoin': 'https://www.reddit.com/r/Bitcoin/.rss?format=xml',
    'Bitcoin.org': 'https://bitcoin.org/en/rss/blog.xml',
    'NewsBTC': 'http://www.newsbtc.com/feed/',
    'SpectroCoin-Blog': 'https://blog.spectrocoin.com/en/feed/',
    'Coinbase-Blog': 'https://blog.coinbase.com/feed',
    'Blocktech.io': 'http://blocktech.io/rssfeed/',
    'ETH-News': 'https://www.ethnews.com/rss.xml'   
    }

last_timestamp = datetime.now() - timedelta(days = 364)

while True:
    news_df = pd.DataFrame(columns = ['timestamp', 'feed', 'title', 'tags', 'link', 'author'])

    for i in range(216):
        feed_json = []
        temp_df = pd.DataFrame(columns = ['timestamp', 'feed', 'title', 'tags', 'link', 'author'])
        count = 0

        for feed, url in feeds.items():
            d = feedparser.parse(url)
            feed_json.append(d)

        for entries in feed_json:
            try:
                feed_name = entries['feed']['generator']
                for entry in entries['entries']:
                    entry_title = entry['title']
                    entry_tags = entry['tags']
                    entry_link = entry['link']
                    try:
                        entry_author = entry['author']
                    except: 
                        entry_author = 'none'
                    entry_time = entry['published_parsed']
                    entry_time = datetime.fromtimestamp(time.mktime(entry_time))

                    temp_df.loc[count] = [entry_time, feed_name, entry_title, entry_tags, entry_link, entry_author]

                    count += 1
            except:
                pass
        temp_df = temp_df.loc[temp_df['timestamp'] > last_timestamp]
        temp_df.sort_values('timestamp', inplace = True)

        timestamp_ls = temp_df['timestamp'].values
        title_ls = temp_df['title'].values
        link_ls = temp_df['link'].values

        for i in range(len(timestamp_ls)):
            print(timestamp_ls[i], title_ls[i], link_ls[i])
        
        
        news_df = news_df.append(temp_df)
        news_df.sort_values('timestamp', inplace = True)
        news_df.reset_index(inplace = True, drop = True)
        news_df.drop_duplicates('timestamp', keep = 'first')
        last_timestamp = news_df['timestamp'].loc[len(news_df) - 1]


        time.sleep(400)
    
    now = datetime.now()
    now_str = datetime.strftime(now, '%d_%m_%Y_%H_%M')
    filename = 'crypto_news_' + now_str
    news_df.to_csv(filename, index = False)    
```
