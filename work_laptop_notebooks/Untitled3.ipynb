```python
from itertools import combinations
import pandas as pd
import numpy as np
from collections import Counter
pd.set_option('max_colwidth', 150)
```

```python
player_df = pd.read_excel('D:/data/memorial_player_ls.xlsx')
player_ls = player_df['player'].values
salary_ls = player_df['salary'].values
odds_ls = player_df['odds'].values
```

```python
combos = combinations(player_ls, 6)
```

```python
salary_dict = dict(zip(player_ls, salary_ls))

valid_ls = []
for combo in combos:
    salary_total = 0
    for player in combo:
        salary_total += salary_dict[player]
        
    if salary_total > 47000 and salary_total <= 50000:
        valid_ls.append(combo)
        
print(len(valid_ls))
```

```python
def calc_mean_odds(ls):
    total_odds = 0
    for player in ls:
        total_odds += odds_dict[player]
        
    return total_odds / 6

odds_dict = dict(zip(player_ls, odds_ls))

lineup_df = pd.DataFrame({'lineup': valid_ls})
lineup_df['avg_odds'] = lineup_df['lineup'].apply(lambda x: calc_mean_odds(x))

lineup_df.sort_values('avg_odds', inplace = True)
lineup_df = lineup_df.head(120)
```

```python
lineup_df
```

```python
combos_3_ls = []
for index, row in lineup_df.iterrows():
    combos_3 = combinations(row['lineup'], 3)
    
    for combo in combos_3:
        combos_3_ls.append(sorted(combo))
    
```

```python
def combine_str(ls):
    return ls[0] +', '+ ls[1] +', '+ ls[2]

count_df = pd.DataFrame({'3_man_combos': combos_3_ls})
count_df['name_string'] = count_df['3_man_combos'].apply(lambda x: combine_str(x))

grouped_df = count_df.groupby('name_string').count()
```

```python
grouped_df.reset_index(inplace = True)
grouped_df.sort_values('3_man_combos', inplace = True, ascending = False)
```
