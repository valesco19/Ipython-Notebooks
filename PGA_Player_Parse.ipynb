```python
os.chdir('/home/valesco/Datasets/PGA_Data/PGA/')

#Read strokes gained csv
strokes_gained_df = pd.read_csv('/home/valesco/Datasets/PGA_Data/PGA_strokes_gained.csv', sep = '\t')


col_list = ['player_id', 'last_name', 'first_name', 'date_started', 'year', 'tournament_id', 'course_num', 'lasers', 'round_id',
                'round_num', 'hole_num', 'hole_num_played', 'hole_id', 'strokes_hole',
                'vs_par_day', 'vs_par_tournament']

for z in range(12):
    asc_col = 'asc_' + str(z + 1)
    club_col = 'club_' + str(z + 1)
    con_col = 'con_' + str(z + 1)
    dist_of_shot_col = 'dist_of_shot_' + str(z + 1)
    start_loc_col = 'start_loc_col_' + str(z + 1)
    dist_left_to_hole_col = 'dist_left_to_hole_' + str(z + 1)
    stroke_num_col = 'stroke_num_' + str(z + 1)
    pid_col = 'pid_' + str(z + 1)
    prv_col = 'prv_' + str(z + 1)
    putt_num_col = 'put_num_' + str(z + 1)
    shot_text_col = 'shot_text_' + str(z + 1)
    t_col = 't_' + str(z + 1)
    from_tee_col = 'from_tee_' + str(z + 1)
    time_col = 'time_' + str(z + 1)
    end_loc_col = 'end_loc' + str(z + 1)
    x_coord_col = 'x_coord_' + str(z + 1)
    y_coord_col = 'y_coord_' + str(z + 1)
    z_coord_col = 'z_coord_' + str(z + 1)

    col_list.extend([asc_col, club_col, con_col, dist_of_shot_col, 
                    start_loc_col, dist_left_to_hole_col, stroke_num_col, pid_col, 
                    prv_col, putt_num_col, shot_text_col, t_col, from_tee_col, 
                    time_col, end_loc_col, x_coord_col, y_coord_col, z_coord_col])

#loop through tournament folders
os.chdir('/home/valesco/Datasets/PGA_Data/PGA_Dir/')
tournaments = glob.glob('*')


for tourny_prefix in tournaments:
    #try:
    print('starting {}'.format(tourny_prefix))
    
    tourny_path = '/home/valesco/Datasets/PGA_Data/PGA/' + tourny_prefix
    os.chdir(tourny_path)


    setup_json = open(tourny_prefix + '_setup.json')
    setup_data = json.load(setup_json)

    course_json = open(tourny_prefix + '_course.json')
    course_data = json.load(course_json)
    
    tee_time_json = open('{}_teetimes.json'.format(tourny_prefix))
    tee_time_data = json.load(tee_time_json)

    course_dict = {}

    for c, course in enumerate(tee_time_data['tournament']['rounds'][0]['courses']):
        course_num = course['CourseNumber']
        course_dict[course_num] = {}

        course_dict[course_num]['course_name'] = course['CourseName']
        course_dict[course_num]['lasers'] = course['holes'][0]['IsCoordinateValid']

    player_round_data = pd.read_csv('/home/valesco/Datasets/PGA_Data/PGA_Dir/{}/{}_player_round_data.csv'.format(tourny_prefix, tourny_prefix))

    player_pool = {}

    for p, field in enumerate(setup_data['trn']['field']):
        player_id = field['id']
        player_pool[player_id] = {}
        player_pool[player_id]['last_name'] = field['name']['last'] 
        player_pool[player_id]['first_name'] = field['name']['first']

    holes_df = pd.DataFrame(columns = col_list)


    for p in tqdm(player_pool.keys()):
        #try:
        p_str = 'PlayerScoreCards/' + p + '.json'
        player_json = open(p_str)
        player_data = json.load(player_json)

        player_id = player_data['p']['id']
        first_name = player_pool[p]['first_name']
        last_name = player_pool[p]['last_name']
        tournament_id = course_data['permNum']
        tournament_year = tourny_prefix.split('_')[1]
        
        try:
            for rnd, rnds in enumerate(player_data['p']['rnds']):
                round_str = str(rnd + 1)
                course_num = str(player_round_data['course_num'].loc[(player_round_data['player_id'] == int(player_id)) & (player_round_data['round'] == int(round_str)) &
                      (player_round_data['year'] == int(tournament_year))].values[0])

                course_lasers = course_dict[course_num]['lasers']

                month = setup_data['trn']['rnds'][rnd]['courses'][0]['teetimes'][0]['groups'][0]['start']['dateMM']
                day = setup_data['trn']['rnds'][rnd]['courses'][0]['teetimes'][0]['groups'][0]['start']['dateDD']
                year = setup_data['trn']['rnds'][rnd]['courses'][0]['teetimes'][0]['groups'][0]['start']['dateYYYY']
                date = '{}_{}_{}'.format(month, day, year)

                round_id = player_id + '_' + tournament_year + '_' + tournament_id + '_' + course_num + '_' + round_str
                for hole, holes in enumerate(rnds['holes']):

                    hole_num = str(holes['cNum'])
                    hole_of_day = str(holes['n'])
                    hole_id = tournament_year + '_' + tournament_id + '_' + course_num + '_' + round_str + '_' + hole_num
                    strokes_on_hole_str = str(holes['sc'])
                    vs_par_day = str(holes['pDay'])
                    vs_par_total = str(holes['pTot'])

                    temp_hole_df = pd.DataFrame(columns = ['player_id', 'last_name', 'first_name', 'date_started', 'year',
                                'tournament_id', 'course_num', 'lasers', 'round_id',
                                'round_num', 'hole_num', 'hole_num_played', 'hole_id', 'strokes_hole',
                                'vs_par_day', 'vs_par_tournament'])

                    temp_hole_df.loc[0] = [player_id, last_name, first_name, date, int(tournament_year), 
                                           tournament_id, course_num, course_lasers, round_id, 
                                           round_str,hole_num, hole_of_day, hole_id, strokes_on_hole_str, vs_par_day,
                                           vs_par_total]

                    shots_df = pd.DataFrame()

                    for shot, shots in enumerate(holes['shots']):
                        asc_col = 'asc_' + str(shot + 1)
                        club_col = 'club_' + str(shot + 1)
                        con_col = 'con_' + str(shot + 1)
                        dist_of_shot_col = 'dist_of_shot_' + str(shot + 1)
                        start_loc_col = 'start_loc_col_' + str(shot + 1)
                        dist_left_to_hole_col = 'dist_left_to_hole_' + str(shot + 1)
                        stroke_num_col = 'stroke_num_' + str(shot + 1)
                        pid_col = 'pid_' + str(shot + 1)
                        prv_col = 'prv_' + str(shot + 1)
                        putt_num_col = 'put_num_' + str(shot + 1)
                        shot_text_col = 'shot_text_' + str(shot + 1)
                        t_col = 't_' + str(shot + 1)
                        from_tee_col = 'from_tee_' + str(shot + 1)
                        time_col = 'time_' + str(shot + 1)
                        end_loc_col = 'end_loc' + str(shot + 1)
                        x_coord_col = 'x_coord_' + str(shot + 1)
                        y_coord_col = 'y_coord_' + str(shot + 1)
                        z_coord_col = 'z_coord_' + str(shot + 1)

                        shot_df = pd.DataFrame(columns = [asc_col, club_col, con_col, dist_of_shot_col, 
                                start_loc_col, dist_left_to_hole_col, stroke_num_col, pid_col, 
                                prv_col, putt_num_col, shot_text_col, t_col, from_tee_col, 
                                time_col, end_loc_col, x_coord_col, y_coord_col, z_coord_col])


                        asc = shots['asc']
                        club = shots['club']
                        con = shots['con']
                        dist_of_shot = shots['dist']
                        start_loc = shots['from']
                        dist_left_to_hole = shots['left']
                        stroke_num = shots['n']
                        pid = shots['pid']
                        prv = shots['prv']
                        putt_num = shots['putt']
                        shot_text = shots['shotText']
                        t = shots['t']
                        from_tee = shots['tee']
                        time_ = shots['time']
                        end_loc = shots['to']
                        x_coord = shots['x']
                        y_coord = shots['y']
                        z_coord = shots['z']

                        shot_df.loc[0] = [asc, club, con, dist_of_shot, start_loc, dist_left_to_hole,
                            stroke_num, pid, prv, putt_num, shot_text, t, from_tee, time_, end_loc,
                            x_coord, y_coord, z_coord]

                        shots_df = pd.concat([shots_df, shot_df], axis = 1)

                    temp_df = pd.concat([temp_hole_df, shots_df], axis = 1)

                    holes_df = pd.concat([holes_df, temp_df], axis = 0, ignore_index = True)
        except:
            pass
        
    holes_df = holes_df[col_list]

    base_path = '/home/valesco/Datasets/PGA_Data/PGA_Dir/' + tourny_prefix
    os.chdir(base_path)


    holes_df.to_csv(tourny_prefix + '_player_holes.csv', index = False)

    print(tourny_prefix, ' succeeded!!')
    #except Exception as e:
        #print(tourny_prefix, ' failed!!! ', e)
```
