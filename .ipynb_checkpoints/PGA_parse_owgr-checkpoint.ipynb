```python
player_ids = pd.read_csv('/home/valesco/Datasets/PGA_Data/player_id.csv')
player_id_ls = player_ids['player_name'].tolist()
num_players = len(player_ids)

rank_df = pd.DataFrame(columns = ['player_id', 'first_name', 'last_name', 'rank', 'date', 'week'])
rank_count = 0

year_ls = ['2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017']

for year in tqdm(year_ls):
    try:
        for week in range(1,53):
            year_str = year + '_week_' + str(week) 

            pdf_file = open(year_str + '.pdf', 'rb')

            read_pdf = pdf_read(pdf_file)
            num_pages = read_pdf.getNumPages()

            pdf_ls = []
            for i in range(num_pages):
                page = read_pdf.getPage(i)
                text = page.extractText()
                str_ls = text.split('\n')
                pdf_ls.extend(str_ls)

                if i == 0:
                    date = pdf_ls[-5]
                    date_stamp = datetime.strptime(date, '%d %B %Y')


            for l, line in enumerate(pdf_ls):
                try:
                    if line[0] == '<':
                        rank = pdf_ls[l - 2]
                        name = pdf_ls[l + 1]
                        player_id_result = process.extractOne(name, player_id_ls)
                        if player_id_result[1] > 89:
                            p_id = int(player_ids['player_id'].loc[player_ids['player_name'] == player_id_result[0]].values[0])
                            player_name = player_id_result[0]
                            first_name = player_name.split(',')[1]
                            first_name = first_name.replace(' ', '')
                            last_name = player_name.split(',')[0]
                        else:
                            p_id = randint(66666,99999)
                            first_name = name.split(' ')[0]
                            last_name = name.split(' ')[1]
                            num_players += 1
                            player_id_name = last_name + ', ' + first_name
                            player_ids.loc[num_players] = [p_id, player_id_name]
                            player_id_ls = player_ids['player_name'].tolist()
                        rank_df.loc[rank_count] = [int(p_id),first_name, last_name, rank, date_stamp, year_str]
                        rank_count += 1
                except:
                    pass
    except:
        pass
```

```python
os.chdir('/home/valesco/Datasets/PGA_Data/pga_master_files/')
rank_df.to_csv('owgr_final.csv', sep = ';', index = False)
```
