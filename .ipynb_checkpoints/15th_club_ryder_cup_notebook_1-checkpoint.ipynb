```python
os.chdir('/home/valesco/Datasets/PGA_Data/15th Club/')
ryder_df = pd.read_csv('hackathon-data-rc.csv')

def true_false_dummy(x):
    if x == 1:
        return 1
    else:
        return 0
    
    
def yards_to_inch(x):
    return int(x) * 36


def create_match_id(df):
    match_title = df['mp_name'].split(' ')
    day_of_week = match_title[0]
    match_type = match_title[1]
    match_num = match_title[-1]
    
    day_dict = {'Friday': 'fri', 'Saturday': 'sat', 'Sunday': 'sun'}
    match_type_dict =  {'Foursomes': 'fs', 'Fourballs': 'fb', 'Singles': 'sing'}
    
    match_id = day_dict[day_of_week] + '_' + match_type_dict[match_type] + '_' + match_num
    
    return(match_id)

def apply_penalty(df):
    player = df['player_shot_name']
    match_id = df['match_id']
    hole_num = df['hole_no']  
    next_shot = df['shot_no'] + 1
    
    temp_row = ryder_df.loc[(ryder_df['match_id'] == match_id) & (ryder_df['hole_no'] == hole_num) &
                         (ryder_df['shot_no'] == next_shot) & (ryder_df['player_shot_name'] == player)]
    if len(temp_row) > 0:
        if temp_row['penalty'].iloc[0] == 1:
            return 1
        else:
            return 0
    else:
        return 0

ryder_df['match_id'] = ryder_df.apply(create_match_id, axis = 1)
ryder_df['penalty_this_shot'] = ryder_df.apply(apply_penalty, axis = 1)
ryder_df['penalty'] = ryder_df['penalty'].apply(lambda x: true_false_dummy(x))
ryder_df['completed'] = ryder_df['completed'].apply(lambda x: true_false_dummy(x))
ryder_df['putt'] = ryder_df['putt'].apply(lambda x: true_false_dummy(x))
ryder_df['hole_dist'] = ryder_df['yards'].apply(lambda x: yards_to_inch(x))
ryder_df.replace({'Rafa Cabrera Bello': 'Rafa CabreraBello', 'J B Holmes': 'J.B. Holmes',
                  'Rory Mcilroy': 'Rory McIlroy', 'Greenside Bunker': 'Sand', 'Green Fringe': 'Fringe', 
                  'Intermediate Rough': 'Intermediate','Fairway Bunker': 'Sand'}, inplace = True)

ryder_df.tail()
```

```python
def myround(x, base):
    return (int(base * round(float(x) / base)))

def calc_sg(df):
    if df['completed'] == 0:
        return 0, np.nan, np.nan
    else:
        start_loc = df['lie_before']
        end_loc = df['lie_after']
        start_shot_dist = df['left_to_pin_before']
        end_shot_dist = df['left_to_pin_after']

        penalty = 0
        
        if df['penalty_this_shot'] == 1:
            penalty = 1
            if end_loc == 'Out of Bounds':
                end_shot_dist = start_shot_dist
                end_loc = start_loc
            else:
                team = df['shot_team']
                match = df['mp_name']
                hole = df['hole_no']
                shot = df['shot_no'] + 1

                end_shot_dist = ryder_df['left_to_pin_after'].loc[(ryder_df['shot_team'] == team) & 
                                    (ryder_df['shot_no'] == shot) & (ryder_df['mp_name'] == match) & 
                                    (ryder_df['lie_before'] == 'Water') & (ryder_df['hole_no'] == hole)].values[0]

                end_loc = ryder_df['lie_after'].loc[(ryder_df['shot_team'] == team) & (ryder_df['shot_no'] == shot) &
                                    (ryder_df['mp_name'] == match) & (ryder_df['lie_before'] == 'Water') &
                                    (ryder_df['hole_no'] == hole)].values[0]        


        if start_loc == 'Green' or start_loc == 'Fringe':
            start_dist_rounded = myround(start_shot_dist / 12, 1)
            start_dist_col = 'PuttDistinfeet'
            
            if start_dist_rounded > 112:
                start_dist_rounded = 112
            if start_dist_rounded < 1:
                start_dist_rounded = 1
                
        elif start_loc in ['Tee', 'Fairway', 'Intermediate', 'Rough', 'Sand']:
            start_dist_rounded = myround(start_shot_dist / 36, 20)
            start_dist_col = 'Distance'
            
            if start_dist_rounded < 20:
                start_dist_rounded = 20
        else:
            start_dist_rounded = 0

        if end_loc == 'Green' or end_loc == 'Fringe':
            end_dist_rounded = myround(end_shot_dist / 12, 1)
            end_dist_col = 'PuttDistinfeet'
            
            if end_dist_rounded > 112:
                end_dist_rounded = 112
                
            if end_dist_rounded < 1 and end_shot_dist > 0:
                end_dist_rounded = 1
            
        elif end_loc in ['Tee', 'Fairway', 'Intermediate', 'Rough', 'Sand']:
            try:
                end_dist_rounded = myround(end_shot_dist / 36, 20) 
                end_dist_col = 'Distance'
                
                if end_dist_rounded < 20:
                    end_dist_rounded = 20
            except:
                return 0, np.nan, np.nan
        else:
            end_dist_rounded = 0
        
        if start_dist_rounded <= 2 and start_loc == 'Green':
            end_dist_rounded = 0

        if start_dist_rounded == 0:
            return 0, np.nan, np.nan
        else:
            start_sg = strokes_gained[start_loc].loc[strokes_gained[start_dist_col] == start_dist_rounded].values[0]

        if end_dist_rounded == 0:
            end_sg = 0
        else:
            end_sg = strokes_gained[end_loc].loc[strokes_gained[end_dist_col] == end_dist_rounded].values[0]
        
        return (start_sg - end_sg - 1 - penalty), start_loc, start_dist_rounded
```

```python
ryder_df['stroke_gained'] = ryder_df.apply(calc_sg, axis = 1)
```
