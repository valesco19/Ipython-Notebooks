```python
import json
import os
import glob
import pandas as pd
import requests
import time
import urllib
from random import randint
from bs4 import BeautifulSoup as bs
from tqdm import tqdm
from datetime import datetime
```

```python
os.chdir('/home/valesco/Datasets/PGA_Data/PGA_Stats/2017/')

base_url = 'http://pgatour.com/data/r/stats/2017/'

#soup = bs(requests.get(base_url).text)
stat_list_json = open('2017_stats_dict.json')
stat_list = json.load(stat_list_json)

stat_dict = {}
for key, stat in tqdm(stat_list.items()):
    #try:
    stat_url = base_url + stat + '.json'
    response = requests.get(stat_url)
    data = json.loads(response.text)

    if len(data['tours'][0]['years'][0]['stats'][0]['details']) > 4:
        dict_key = data['tours'][0]['years'][0]['stats'][0]['statID']
        stat_dict[dict_key] = data['tours'][0]['years'][0]

    time.sleep(1)
    #except:
        #pass

now = datetime.now()
date_str = now.strftime('%m_%d_%Y')

json_name = date_str + '_stats.json'
with open(json_name, 'w') as outfile:
            json.dump(stat_dict, outfile)
```
