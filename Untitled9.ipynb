```python
test_url = 'http://www.footballdb.com/players/drew-brees-breesdr01'

browser = webdriver.Chrome()

browser.get(test_url)

player_name = browser.find_element_by_xpath('//*[@id="playertop"]/h1').text

banner = browser.find_element_by_xpath('//*[@id="playerbanner"]')

banner_text = banner.text

banner_ls = banner_text.split('\n')

player_team = banner_ls[0].split(',')[0]
player_num = banner_ls[0].split(',')[1]
player_num = player_num.replace('#','')
player_pos = banner_ls[1].split(':')[1]
player_bday = banner_ls[2].split(':')[1]
player_coll = banner_ls[3].split(':')[1]

if len(banner_ls) > 4:
    text = banner_ls[4].split(':')[1]
    text = text.split('(')[1]
    text = text.split(')')[0]
    player_draft = re.findall(r'\d+', text)[0]
    
    

print('{} plays for {} at {} born {} played college at {} and drafted {} and wears #{}'.format(player_name, player_team,
                                                     player_pos, player_bday, player_coll, player_draft, player_num))


#game_log = browser.find_element_by_xpath('//*[@id="playerlinks"]/a[3]')
game_log = browser.find_element_by_xpath('//a[contains(@href,"gamelogs")]')
game_log.click()

game_log_year = browser.find_element_by_class_name('linkset hidden-xs')

game_log_ls = game_log_year.find_elements_by_css_selector('a')

for year in game_log_ls:
    print(year.text)



browser.close()
```
