```python
while True:
    sports = ['PGA', 'NFL', 'NBA']
    
    for sport in sports:
        try:
            contest_dict = {}
            username = 'nemisis'
            pword = 'degen2345'

            browser = webdriver.Chrome()
            url = 'https://www.draftkings.com/lobby#/{}/0/All'.format(sport)

            browser.get(url)
            time.sleep(randint(2,5))

            browser.maximize_window()

            login = browser.find_element_by_xpath('//input[@placeholder="Username or Email"]')
            login.send_keys(username + Keys.TAB)

            time.sleep(1)

            password_text = browser.find_element_by_xpath('//input[@placeholder="Password"]')
            password_text.send_keys(pword)

            time.sleep(randint(2,3))

            login_btn = browser.find_element_by_xpath('//*[contains(text(), "LOG IN")]')
            login_btn.click()

            time.sleep(randint(2,3))

            for e, r in enumerate(range(0)):
                dk_header = browser.find_element_by_xpath('//div[@id="dkjs-subnav"]')

                sport_btn = dk_header.find_element_by_link_text(sport)
                sport_btn.click()

                time.sleep(randint(2,5))

                #live_icon = browser.find_element_by_xpath('//*[contains(text(), "View Live Contests")]')
                #live_icon = browser.find_element_by_xpath('//span[@class="icon-live"]')
                #live_icon.click()

                live_icon = browser.find_element_by_xpath('//*[contains(text(), "View Live Contests")]')
                browser.execute_script('return arguments[0].scrollIntoView(true);', live_icon)
                #live_icon = browser.find_element_by_xpath('//span[@class="icon-live"]')
                browser.execute_script('arguments[0].click();', live_icon)
                #live_icon.click()

                time.sleep(randint(3,5))

                game_links = browser.find_elements_by_xpath('//a[@href]')
                contest_ls = []

                for game in game_links:
                    if '/contest/gamecenter/' in game.get_attribute('href'):
                        contest_ls.append(game)

                browser.execute_script("return arguments[0].scrollIntoView(true);", contest_ls[e])
                time.sleep(1)
                contest_ls[e].click()

                time.sleep(randint(2,4))

                contest_id = browser.current_url.split('/')[-1]

                contest_details_link = browser.find_element_by_id('contest-details-link')
                contest_details_link.click()

                time.sleep(randint(1,2))

                contest_details = browser.find_element_by_id('fancybox-content')

                contest_text = contest_details.text

                contest_dict[contest_id] = contest_text

                time.sleep(randint(1,2))

                close_btn = browser.find_element_by_id('fancybox-close')
                close_btn.click()

                time.sleep(randint(1,2))

                csv_btn = browser.find_element_by_id('export-lineups-csv')
                csv_btn.click()

                time.sleep(randint(1,2))

                browser.execute_script('window.history.go(-1)')

                time.sleep(randint(1,2))
        except:
            now = datetime.now()
            print('{} failed at {}'.format(sport, now))
            pass


        now = datetime.now()
        now_str = now.strftime('%Y_%m_%d_%H_%M')
        filename = now_str + '_{}_contest_details.json'.format(sport)

        with open(filename, 'w') as outfile:
            json.dump(contest_dict, outfile)

        print(now_str + '{} Success!!!'.format(sport))


    time.sleep(40000)
```
