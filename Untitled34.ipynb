```python
def myround(x, base):
    return (int(base * round(float(x) / base)))

def return_previous_tourny_info(tid):
    sql_select = 'SELECT * FROM pga.tournament_info where tournament_num = {} order by tournament_year desc'.format(tid)
    tourny_df = pd.read_sql(sql_select, engine)
    
    return tourny_df

def return_previous_course_holes(tid):
    sql_select = 'SELECT * FROM pga.course_holes where tournament_id = {}  order by `year` desc, round_num asc, hole_num asc'.format(tid)
    
    course_holes = pd.read_sql(sql_select, engine)
    
    return course_holes


def return_previous_sg(tid):
    sql_select = 'SELECT * FROM pga.strokes_gained where tournament_id = {} order by `year` asc'.format(tid)
    
    strokes_gained = pd.read_sql(sql_select, engine)
    return strokes_gained 

def baseline_course_sg():
    sql_select = 'SELECT * FROM pga.strokes_gained'
    
    base_sg = pd.read_sql(sql_select, engine)
    
    #base_sg['dist_bin'] = base_sg['shot_dist'].apply(lambda x: myround(x / 36, 40))
    
    grouped_df = base_sg[['tournament_id', 'year', 'stroke_gained', 'shot_loc']].groupby(
                        ['tournament_id', 'year', 'shot_loc']).mean()
    
    grouped_df.reset_index(inplace = True)
    grouped_df.sort_values(['tournament_id', 'year', 'shot_loc'], inplace = True)
    
    return grouped_df
    
course_holes_df = return_previous_course_holes(tid)
sg_df = return_previous_sg(tid)
grouped_df = baseline_course_sg()
grouped_df
```

```python
def count_shot_loc():
    sql_select = 'SELECT * FROM pga.strokes_gained'
    
    base_sg = pd.read_sql(sql_select, engine)
    base_sg = base_sg.loc[(base_sg['shot_loc'] != 'Green') | (base_sg['shot_loc'] != 'Intermediate')]
    base_sg['dist_bin'] = base_sg['shot_dist'].apply(lambda x: myround(x / 36, 20))
    
    grouped_df = base_sg[['tournament_id', 'year', 'stroke_gained', 'shot_loc']].groupby(
                        ['tournament_id', 'year', 'shot_loc']).count()
    
    grouped_df.reset_index(inplace = True)
    grouped_df.sort_values(['tournament_id', 'year', 'shot_loc'], inplace = True)
    
    return grouped_df

count_df = count_shot_loc()
```

```python
count_df
```

```python
sum_df = count_df[['tournament_id', 'year', 'stroke_gained']].groupby(['tournament_id', 'year']).sum()
sum_df.reset_index(inplace = True)
sum_df
```

```python
def shot_count_percent(row):
    tid = row['tournament_id']
    year = row['year']
    
    total = sum_df['stroke_gained'].loc[(sum_df['tournament_id'] == tid) & (sum_df['year'] == year)].values[0]
    
    return row['stroke_gained'] / total

count_df['percent_shots'] = count_df.apply(shot_count_percent, axis = 1)

count_df
```

```python
fairway_df = count_df.loc[count_df['shot_loc'] == 'Fairway']
fairway_df['rank'] = fairway_df['percent_shots'].rank()
fairway_df.sort_values('rank', inplace = True)

tee_df = count_df.loc[count_df['shot_loc'] == 'Tee']
tee_df['rank'] = tee_df['percent_shots'].rank()
tee_df.sort_values('rank', inplace = True)

green_df = count_df.loc[(count_df['shot_loc'] == 'Green')]
green_df['rank'] = green_df['percent_shots'].rank()
green_df.sort_values('rank', inplace = True)

rough_df = count_df.loc[(count_df['shot_loc'] == 'Rough')]
rough_df['rank'] = rough_df['percent_shots'].rank()
rough_df.sort_values('rank', inplace = True)

count_rank_df = pd.concat([fairway_df, tee_df, green_df, rough_df], axis = 0)
count_rank_df 
```
