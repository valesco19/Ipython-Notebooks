```python
sql = 'SELECT * FROM pga.player_holes WHERE tournament_id = 12;'

round_df = pd.read_sql(sql, con = conn)
```

```python
#round_df['year'] = round_df['hole_id'].apply(lambda x: x.split('_')[0])
round_df['full_name'] = round_df['first_name'] + ' ' + round_df['last_name']
round_df.head()
```

```python
sg_df = pd.DataFrame(columns = ['full_name', 'player_id', 'shot_sg', 'shot_loc', 'shot_start_dist', 'year'])

for i in range(1,6):
    sg_col = 'shot_{}_sg'.format(i)
    sg_loc_col = 'shot_{}_loc'.format(i)
    sg_dist_col = 'shot_{}_start_dist'.format(i)
    
    temp_df = round_df[['full_name', 'player_id', sg_col, sg_loc_col, sg_dist_col, 'year']]
    temp_df.columns = ['full_name', 'player_id', 'shot_sg', 'shot_loc', 'shot_start_dist', 'year']
    sg_df = sg_df.append(temp_df)
```

```python
def myround(x, base):
    return (int(base * round(float(x) / base)))

def round_dist(df):
    if df['shot_loc'] == 'Green' or df['shot_loc'] == 'Fringe':
        temp_dist = myround(df['shot_yards'], 1)
        if temp_dist < 1:
            temp_dist = 1
        return temp_dist
    elif df['shot_loc'] == '':
        return 0
    else:
        temp_dist = myround(df['shot_yards'], 20)
        if temp_dist < 20:
            temp_dist = 20
        return temp_dist

sg_df['shot_yards'] = sg_df['shot_start_dist'] / 36
sg_df['shot_bin'] = sg_df.apply(round_dist, axis = 1)
sg_df
```

```python
agg_dict = {'shot_sg': 'mean', 'shot_start_dist': 'mean', 'player_id': 'count'}
player_grouped = sg_df.groupby(['full_name', 'shot_loc']).agg(agg_dict)
```

```python
#player_grouped.reset_index(inplace = True)
player_grouped.loc[player_grouped['shot_loc'] == 'Fairway'].sort_values(by = 'shot_sg', ascending = False)
```
