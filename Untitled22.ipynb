```python
import json
import os
import glob
import pandas as pd
import requests
import time
import urllib
from random import randint
from bs4 import BeautifulSoup as bs
from tqdm import tqdm
from datetime import datetime
```

```python
base_url = 'http://data.pga.com/jsonp/sasd/rydercup/2016/scorecard/'

ryder_cup_dict = {}

for i in range(175,303):
    final_url = base_url + str(i) + '.json'
    time.sleep(1)

    response = requests.get(final_url)
    
    if response.status_code == 200:
        string = response.text
        string = string.replace('callbackWrapper(', '')[:-2]
        data = json.loads(string)
        ryder_cup_dict[str(i)] = data
        print(str(i) + '.json succeeded')
    else:
        print(response.status_code)  
    
```

```python
ryder_cup_results_df = pd.DataFrame(columns = ['match_title', 'match_type', 'match_displayno', 'tee_time',
                                               'team_won', 'win_score','team_id', 'player_name', 'player_id', 
                                               'player_country','hole_1_score', 'hole_2_score','hole_3_score','hole_4_score',
                                               'hole_5_score', 'hole_6_score','hole_7_score','hole_8_score',
                                               'hole_9_score','hole_10_score','hole_11_score','hole_12_score',
                                               'hole_13_score','hole_14_score','hole_15_score','hole_16_score',
                                               'hole_17_score','hole_18_score'])

count = 0
for key in ryder_cup_dict.keys():
    match_title = ryder_cup_dict[key]['matches']['sessiontype']
    match_type = ryder_cup_dict[key]['matches']['type']
    match_teetime = ryder_cup_dict[key]['matches']['teetime']
    match_displayno = ryder_cup_dict[key]['matches']['displayno']
    match_team_won = ryder_cup_dict[key]['matches']['leader']
    match_win_score = ryder_cup_dict[key]['matches']['score']
    
    for team in ryder_cup_dict[key]['matches']['team']:
        country = team['name']
        team_id = team['id']
        if match_type == 'singles':
            player_name = team['player']['fname'] + ' ' + team['player']['lname']
            player_id = team['player']['id']
            scorecard = team['player']['scoring']
            
            temp_col_ls = [match_title, match_type, match_displayno, match_teetime, match_team_won,
                            match_win_score, team_id, player_name, player_id, country]

            temp_col_ls.extend(scorecard)

            ryder_cup_results_df.loc[count] = temp_col_ls

            count += 1
    
        else:
            for player in team['player']:
                player_name = player['fname'] + ' ' + player['lname']
                player_id = player['id']
                scorecard = player['scoring']

                temp_col_ls = [match_title, match_type, match_displayno, match_teetime, match_team_won,
                            match_win_score, team_id, player_name, player_id, country]

                temp_col_ls.extend(scorecard)

                ryder_cup_results_df.loc[count] = temp_col_ls

                count += 1
    
```

```python
ryder_cup_results_df.to_csv('/home/valesco/Datasets/PGA_Data/15th Club/ryder_cup_results.csv', index = False)
```

```python
def create_match_id(df):
    match_title = df['match_title']
    day_of_week = match_title.split(' ')[0]
    match_type = df['match_type']
    match_num = df['match_displayno']
    
    day_dict = {'Friday': 'fri', 'Saturday': 'sat', 'Sunday': 'sun'}
    match_type_dict =  {'foursomes': 'fs', 'fourball': 'fb', 'singles': 'sing'}
    
    match_id = day_dict[day_of_week] + '_' + match_type_dict[match_type] + '_' + str(match_num)
    
    return(match_id)

ryder_cup_results_df['match_id'] = ryder_cup_results_df.apply(create_match_id, axis = 1)
ryder_cup_results_df
```
