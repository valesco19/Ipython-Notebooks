```python
def scrape_money_leaders_year(url):
    #Downloads the money winners by year for the PGA tour
    #using Requests
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'lxml')
    money = []
    player = []
    find_num = []
    for item in soup.findAll("td", "hidden-small"):
        find_num.append(item.text)
    for i in range(10):
        try:
            float(find_num[i])
            x = i
            break
        except ValueError:
            pass
    for item in soup.findAll("td", "hidden-small")[1::x]:
        money.append(item.text)
    for item in soup.findAll('td', 'player-name'):
        player.append(item.text)
    player = ([p.strip('\n') for p in player])
    player = ([p.replace(u'\xa0', u' ') for p in player])
    money_list = dict(zip(player, money))
    return money_list
```

```python
money_dict = scrape_money_leaders_year('http://www.pgatour.com/stats/stat.109.2016.html')
money_dict
```

```python
def get_links():
    urls = []
    base_url = "http://www.pgatour.com/stats/stat."
    r = [122,129,130]
    for i in r:
        urls.append((base_url + str(i) + ".html"))
    return urls

def scrape_stats(url):
    #Downloads the stat for 2016 for the PGA tour
    #using Requests
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'lxml')
    head = []
    for item in soup.findAll('h3'):
        head.append(item.text)
    header = head[-1]
    find_num = []
    cat = []
    player = []
    for item in soup.findAll("td", "hidden-small"):
        find_num.append(item.text)
    for i in range(10):
        try:
            float(find_num[i])
            x = i
            break
        except ValueError: 
            pass
    for item in soup.findAll("td", "hidden-small")[1::x]:
        cat.append(item.text)
    for item in soup.findAll('td', 'player-name'):
        player.append(item.text)
    player = ([p.strip('\n') for p in player])
    player = ([p.replace(u'\xa0', u' ') for p in player])
    stat_dict = dict(zip(player, cat))
    return header, stat_dict
```
