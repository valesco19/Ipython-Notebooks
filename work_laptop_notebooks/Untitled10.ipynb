```python
import pandas as pd
import numpy as np
import os
```

```python
qb_df = pd.read_excel('D:/data/qb_matchups.xlsx')
qb_df
```

```python
teams = qb_df.keys()
qb_dict = {}

for team in teams:
    qb_dict[team] = qb_df[team].values
    
combo_df = pd.DataFrame(columns = ['teams', 'week1', 'week2', 'week3', 'week4', 'week5', 'week6', 'week7', 'week8', 
                                    'week9', 'week10', 'week11', 'week12', 'week13', 'week14', 'week15', 'week16'])


count = 0

for team in teams:
    for team1 in teams:
        if team != team1:
            combo_name = team + '/' + team1
            temp_ls = [combo_name]
            
            for i in range(len(qb_dict[team]) - 1):
                if np.isnan(qb_dict[team][i]):
                    temp_ls.append(qb_dict[team1][i])
                elif np.isnan(qb_dict[team1][i]):
                    temp_ls.append(qb_dict[team][i])
                elif qb_dict[team][i] > qb_dict[team1][i]:
                    temp_ls.append(qb_dict[team][i])
                else:
                    temp_ls.append(qb_dict[team1][i])
                    
            combo_df.loc[count] = temp_ls
            count += 1
            
            
            
```

```python
combo_df['total'] = combo_df.sum(axis = 1)
combo_df.sort_values('total', ascending = False)
```
