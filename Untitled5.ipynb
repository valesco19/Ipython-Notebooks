```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
import os
```

```python
url = 'https://www.google.com'

browser = webdriver.Firefox()
browser.get(url)

search = browser.find_element_by_xpath('//*[@id="lst-ib"]')
search.send_keys('new gold rush' + Keys.ENTER)

time.sleep(3)

tools = browser.find_element_by_id('hdtb-tls')
tools.click()

time.sleep(2)

time_btn = browser.find_element_by_xpath('//*[@id="hdtbMenus"]/div/div[2]')
time_btn.click()

past_week = browser.find_element_by_xpath('//*[@id="qdr_w"]/a')
past_week.click()

links = browser.find_elements_by_xpath('//*')
for link in links:
    print(link.get_attribute('href'))

time.sleep(10)
browser.close()
```
