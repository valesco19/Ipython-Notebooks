```python
import json
import os
import glob
import pandas as pd
os.chdir('/home/valesco/Datasets/PGA_Data/PGA/003_2016/')
```

```python
tournaments = glob.glob('*')
```

```python
# Retrieve Player Id's and Player Names from teetimes.json 

playerID_dict = {}
for t in tournaments:
    teetime_string = t + '/' + t + '_teetimes.json'
    try:
        #Load tee time file
        player_data = open(teetime_string)
        data = json.load(player_data)
        
        segment_list = data['tournament']['rounds'][0]['courses'][0]['segments']
        
        for segment in segment_list:
            for group in segment['groups']:
                for player in group['players']:
                    playerID_dict[player['TournamentPlayerID']] = player['LFFull']
    except:
        pass
```

```python
player_df = pd.DataFrame([[key, value] for key,value in playerID_dict.items()], columns = ['player_id', 'player_name'])

player_df.to_csv('player_id.csv')
```
