```python
def fuzz_match(x):
    results = process.extractOne(x, tourny_ls)
    if results[1] > 89:
        return tournament_ids['tournament_id'].loc[tournament_ids['tournament_name'] == results[0]].values[0]
    else:
        return np.nan
    
results_tourny_dict = {}
results_tourny_ls = player_results['tournament_name'].unique()
for results_tourny in results_tourny_ls:
    t_id = fuzz_match(results_tourny)
    results_tourny_dict[results_tourny] = t_id
    

combo_df['tournament_id'] = combo_df['tournament_name'].apply(lambda x: fuzz_match(x))
```

```python
def prev_appear(df):
    player_id = df['player_id']
    tournament_id = df['tournament_id']
    date_cutoff = pd.to_datetime(df['date'])
    temp_df = player_results.loc[(player_results['player_id'] == player_id) & 
                                 (player_results['tournament_id'] == tournament_id) &
                                 (pd.to_datetime(player_results['tournament_start_date']) < date_cutoff)]

    num_appear = temp_df.shape[0]
    if num_appear > 0:
        recent_appear = temp_df['final_rank'].iloc[0].mean()
    else:
        recent_appear = -999

    if num_appear > 2:
        last_3_appear = temp_df['final_rank'].iloc[:3].mean()
    else:
        last_3_appear = -999
        
    if num_appear > 3:
        total_appear = temp_df['final_rank'].mean()
    else:
        total_appear = -999
    return(num_appear, recent_appear, last_3_appear, total_appear)

def recent_results(df):
    try:
        player_id = df['player_id']
        date_cutoff = pd.to_datetime(df['date'])
        temp_df = player_results.loc[(player_results['player_id'] == player_id) & 
                                     (pd.to_datetime(player_results['tournament_start_date']) < date_cutoff)]

        temp_df.sort_values(by = 'tournament_start_date', ascending = False, inplace = True)

        last_result = temp_df['final_rank'].iloc[0].mean()
        last_3_results = temp_df['final_rank'].iloc[:3].mean()
        last_6_results = temp_df['final_rank'].iloc[:6].mean()
        last_10_results = temp_df['final_rank'].iloc[:10].mean()
        last_20_results = temp_df['final_rank'].iloc[:20].mean()
        last_40_results = temp_df['final_rank'].iloc[:40].mean()

        return(last_result, last_3_results, last_6_results, last_10_results, last_20_results, last_40_results)
    except:
        return(-999, -999, -999, -999, -999, -999)
    
#combo_df['tourny_appear'] = combo_df.apply(prev_appear, axis = 1)
#combo_df['recent_results'] = combo_df.apply(recent_results, axis = 1)
shell_ownership['recent_results'] = shell_ownership.apply(recent_results, axis = 1)
shell_ownership['prev_appear'] = shell_ownership.apply(prev_appear, axis = 1)
```

```python
def recent_ownership(df):
    try:
        player_id = df['player_id']
        cutoff_date = pd.to_datetime(df['date'])
        temp_df = combo_df.loc[(combo_df['player_id'] == player_id) & (pd.to_datetime(combo_df['date']) < cutoff_date)]
        temp_df['date'] = pd.to_datetime(temp_df['date'])
        temp_df.sort_values(by = 'date', ascending = False, inplace = True)

        last_own = temp_df['ownership'].iloc[0]
        last_3_own = temp_df['ownership'].iloc[:3].mean()
        last_6_own = temp_df['ownership'].iloc[:6].mean()
        last_10_own = temp_df['ownership'].iloc[:10].mean()

        return (last_own, last_3_own, last_6_own, last_10_own)
    except:
        return (-999, -999, -999, -999)
    
#combo_df['recent_ownership'] = combo_df.apply(recent_ownership, axis = 1)
shell_ownership['recent_ownership'] = shell_ownership.apply(recent_ownership, axis = 1)
```
