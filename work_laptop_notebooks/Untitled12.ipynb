```python
import os
import json

os.chdir('d:/data/d3_apps/distro_dash')
```

```python
json_file = open('10_31_2017_nfl_processed.json')
json_data = json.load(json_file)
json_data
```

```python
sorted_dict = {'identifier': 'PlayerId', 'items': []}
for x in sorted(json_data['items'], key = lambda x: x.get('Salary'), reverse = True):
    sorted_dict['items'].append(x)
    
sorted_dict
```

```python
with open('11_1_2017_nfl_processed.json', 'w') as outfile:
    json.dump(sorted_dict, outfile)
    
```
