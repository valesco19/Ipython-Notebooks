```python
from random import triangular
import random
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def bimodal( low1, high1, mode1, low2, high2, mode2, cutoff):
    toss = random.randint(0, 100)
    if toss >= cutoff:
        return triangular( low1, high1, mode1 ) 
    else:
        return triangular( low2, high2, mode2 )
```

```python

cutmade_ceiling = 80
cutmissed_ceiling = 50

cutmade_percent = 50



def create_distro(cutmade_ceiling, cutmissed_ceiling, cutmade_percent):
    distro = []
    
    cutmade_mode = (cutmade_ceiling - cutmissed_ceiling) / 2 + cutmissed_ceiling
    cutmissed_mode = cutmissed_ceiling / 2
    
    for i in range(200):
        distro.append(round(bimodal(0, cutmissed_ceiling, cutmissed_mode, cutmissed_ceiling, cutmade_ceiling, cutmade_mode,
                               cutmade_percent)))
        
    return distro
```

```python
madecut_ceiling_array = np.linspace(100,70, 125)
missedcut_ceiling_array = np.linspace(40,25, 125)
cutmade_percent_array = np.linspace(90,20, 125)
player_dict = {}

for i in range(len(madecut_ceiling_array)):
    temp_distro = []
    
    jostle = random.uniform(.93, 1.07)
    cutmade_ceiling = madecut_ceiling_array[i] * jostle
    cutmissed_ceiling = missedcut_ceiling_array[i] * jostle
    cutmade_percent = cutmade_percent_array[i] * jostle
    temp_distro.append(create_distro(cutmade_ceiling, cutmissed_ceiling, cutmade_percent))
    player_dict['player{}'.format(i)] = temp_distro
    
player_dict   
```
