```python
url = 'https://www.theopen.com/api/traditional'

data = requests.get(url).json()
data
```

```python
#Retreive Player Scorecards

pid_ls = []

for player in data:
    pid_ls.append(player['ID'])

base_url = 'https://www.theopen.com/api/playerDrawer?playerNo={}'
for pid in pid_ls:
    complete_url = base_url.format(pid)
    
    player_data = requests.get(complete_url).json()
    
    file_name = '{}_round_data.json'.format(pid)
    
    with open(file_name, 'w') as outfile:
        json.dump(player_data, outfile)
        
    time.sleep(.5)
    
```

```python
#Retrieve Tee Times

teetime_url = 'https://www.theopen.com/api/teetimes?day=R{}&page=613EC4C18BD14C62899D21D9454C4464'

for i in range(1,3):
    tee_file_name = 'Round_{}_tee_data.json'.format(i)
    
    complete_url = teetime_url.format(i)
    
    tee_data = requests.get(complete_url).json()
    
    with open(tee_file_name, 'w') as outfile:
        json.dump(tee_data, outfile)
```

```python
#Retrieve stat categories

stat_categories_url = 'https://www.theopen.com/api//playerStatsGroups'

stat_data = requests.get(stat_categories_url).json()

stat_data

stat_ls = []

for cat in stat_data:
    for stat in cat['Stats']:
        stat_ls.append(stat)
        
for stat in stat_ls:
    stat_url = 'https://www.theopen.com/api/playerStats?stats={}'.format(stat)
    
    stat_data = requests.get(stat_url).json()
    
    stat_filename = 'stat_{}.json'.format(stat)
    
    with open(stat_filename, 'w') as outfile:
        json.dump(stat_data, outfile)
        
    time.sleep(.5)
```

```python
#Retrieve hole stats

hole_stat_base_url = 'https://www.theopen.com/api/holestatistics?roundno={}'

for i in range(1,3):
    hole_stat_url = hole_stat_base_url.format(i)
    
    hole_stat_filename = 'Round_{}_hole_stats.json'.format(i)
    
    hole_stat_data = requests.get(hole_stat_url).json()
    
    with open(hole_stat_filename, 'w') as outfile:
        json.dump(hole_stat_data, outfile)
        
```
