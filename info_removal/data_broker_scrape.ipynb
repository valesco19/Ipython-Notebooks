```python
import requests
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup as bs
import time
from datetime import datetime
```

```python
url = 'https://www.privacyrights.org/data-brokers?page=0'

response = requests.get(url, 'lxml').text

soup = bs(response)
soup
```

```python

links = []
terms_of_service = []
privacy_policy = []
opt_outs = []
names = []


for i in range(5):
    url = 'https://www.privacyrights.org/data-brokers?page=' + str(i)
    response = requests.get(url, 'lxml').text

    soup = bs(response)
    rows = soup.select('div[class*="views-row views"]')
    
    for row in rows:
        #find hrefs
        hrefs = row.findAll('a')
        links.append(hrefs[0].attrs['href'])
        names.append(hrefs[0].text)

        try:
            terms_of_service.append(hrefs[1].attrs['href'])
        except:
            terms_of_service.append('null')

        try:
            privacy_policy.append(hrefs[2].attrs['href'])
        except:
            privacy_policy.append('null')

        try:
            opt_out = row.select("div[class*='broker-optout']")
            opt_outs.append(opt_out[0].select('p')[0].text)
        except:
            opt_outs.append('null')
    
    time.sleep(1)
```

```python
broker_df = pd.DataFrame(columns = ['names', 'url', 'terms_of_service', 'privacy_policy', 'opt_out'])

broker_df['names'] = names
broker_df['url'] = links
broker_df['terms_of_service'] = terms_of_service
broker_df['privacy_policy'] = privacy_policy
broker_df['opt_out'] = opt_outs
```

```python
broker_df.to_csv('/home/valesco/Datasets/data_broker_urls.csv', index = False)
```
