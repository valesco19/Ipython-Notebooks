```python
round_robins = {
    1: {'dj': 850, 'kisner': 10000, 'hadwin': 10000, 'wies': 17500}, 
    2: {'kuch': 7500, 'fisher': 10000, 'ikeda': 30000,'zj': 6600},
    3: {'fleetwood': 3300, 'berger': 8000, 'chappell': 8000, 'poulter': 12500},
    4: {'day': 1200, 'oost': 5000, 'duf': 12500, 'hahn': 15000},
    5: {'hideki': 3500, 'cantlay': 4500, 'ca. smith': 10000, 'miyazato': 30000},
    6: {'hatton': 3500, 'hoffman': 8000, 'steele': 10000, 'levy': 12500},
    7: {'noren': 3300, 'finau': 6600, 'pieters': 6600, 'na': 10000},
    8: {'spieth': 1600, 'reed': 2800, 'li': 12500, 'schwartzel': 12500},
    9: {'thomas': 1100, 'molinari': 10000, 'kizzire': 12500, 'list': 7500},
    10: {'perez': 8000, 'woodland': 6600, 'simpson': 8000, 'kim': 15000},
    11: {'casey': 2000, 'fitzpatrick': 10000,'stanley': 12500, 'henley': 12500},
    12: {'sergio': 2800, 'xander': 6600, 'frittelli': 10000, 'sharma': 10000},
    13: {'rory': 750, 'harman': 8000, 'vegas': 17500, 'uihlein': 17500},
    14: {'leishman': 4500, 'grace':6000 , 'bubba': 5000, 'suri': 20000},
    15: {'mickelson': 2500, 'bello': 5000, 'kodaira': 30000, 'howell III': 10000},
    16: {'rahm': 1200, 'aphibarnrat': 8000, 'reavie': 12500, 'keegan': 12500}
}

odds_df = pd.DataFrame(columns = ['player', 'beginning_odds', 'r1_odds', 'group'])
odds_count = 0

for key, value in round_robins.items():
    count = 0
    for k, v in value.items():
        if count == 0:
            p1 = convert_odds(v)
        elif count == 1:
            p2 = convert_odds(v)
        elif count == 2:
            p3 = convert_odds(v)
        elif count == 3:
            p4 = convert_odds(v)
        count += 1
    
    total = p1 + p2 + p3 + p4
    p1_odds = p1 / total
    p2_odds = p2 / total
    p3_odds = p3 / total
    p4_odds = p4 / total
    
    second_count = 0
    for k, v in value.items():
        if second_count == 0:
            odds_df.loc[odds_count] = [k, p1, p1_odds, key]
        elif second_count == 1:
            odds_df.loc[odds_count] = [k, p2, p2_odds, key]
        elif second_count == 2:
            odds_df.loc[odds_count] = [k, p3, p3_odds, key]
        elif second_count == 3:
            odds_df.loc[odds_count] = [k, p4, p4_odds, key]
        
        second_count += 1
        odds_count += 1
    
    
odds_df
```

```python
def select_from_4(dic):
    player_ls = []
    odds_ls = []
    for player, odds in dic.items():
        player_ls.append(player)
        odds_ls.append(odds)
        
    winner = choice(player_ls, 1, odds_ls)
    return winnner

sim_df = pd.DataFrame(columns = ['p1', 'p2', 'p3', 'p4', 'semi_losers'])
sim_count = 0

for a in tqdm(range(5000)):
    final_winners = []
    final_losers = []
    for i in range(4):
        start_num = 4 * i + 1
        group_winners = []
        for j in range(4):
            temp_df = odds_df.loc[odds_df['group'] == start_num + j]
            players_ls = list(temp_df['player'].values)
            odds_ls = list(temp_df['r1_odds'].values)
            winner = choice(players_ls, 1, p = odds_ls)[0]
            odds = odds_ls[players_ls.index(winner)]
            initial_odds = odds_df['beginning_odds'].loc[odds_df['player'] == winner].values[0]

            group_winners.append({'winner': winner, 'odds': initial_odds})

        second_rd_group_1_total = group_winners[0]['odds'] + group_winners[1]['odds']
        second_rd_group_2_total = group_winners[2]['odds'] + group_winners[3]['odds']

        second_rd_player_1_odds = group_winners[0]['odds'] / second_rd_group_1_total
        second_rd_player_2_odds = group_winners[1]['odds'] / second_rd_group_1_total
        second_rd_player_3_odds = group_winners[2]['odds'] / second_rd_group_2_total
        second_rd_player_4_odds = group_winners[3]['odds'] / second_rd_group_2_total

        if second_rd_player_1_odds > .70:
            second_rd_player_1_odds = .70
            second_rd_player_2_odds = .30

        if second_rd_player_2_odds > .70:
            second_rd_player_2_odds = .70
            second_rd_player_1_odds = .30

        if second_rd_player_3_odds > .70:
            second_rd_player_3_odds = .70
            second_rd_player_4_odds = .30

        if second_rd_player_4_odds > .70:
            second_rd_player_4_odds = .70
            second_rd_player_3_odds = .30

        second_rd_winner_1 = choice([group_winners[0]['winner'], group_winners[1]['winner']], 1, p = [second_rd_player_1_odds, second_rd_player_2_odds])[0]
        second_rd_winner_2 = choice([group_winners[2]['winner'], group_winners[3]['winner']], 1, p = [second_rd_player_3_odds, second_rd_player_4_odds])[0]

        for group_winner in group_winners:
            if group_winner['winner'] == second_rd_winner_1:
                second_rd_winner_odds_1 = group_winner['odds']
            if group_winner['winner'] == second_rd_winner_2:
                second_rd_winner_odds_2 = group_winner['odds']

        semi_winners = [{'winner': second_rd_winner_1, 'odds': second_rd_winner_odds_1}, 
                        {'winner': second_rd_winner_2, 'odds': second_rd_winner_odds_2}]

        semi_total = semi_winners[0]['odds'] + semi_winners[1]['odds']

        semi_winner_1_odds = semi_winners[0]['odds'] / semi_total
        semi_winner_2_odds = semi_winners[1]['odds'] / semi_total

        if semi_winner_1_odds > .70:
            semi_winner_1_odds = .70
            semi_winner_2_odds = .30

        if semi_winner_2_odds > .70:
            semi_winner_2_odds = .70
            semi_winner_1_odds = .30

        final_winner = choice([semi_winners[0]['winner'], semi_winners[1]['winner']], 1, [semi_winner_1_odds, semi_winner_2_odds])[0]
        final_winners.append(final_winner)
        
        for semi_winner in semi_winners:
            if semi_winner['winner'] != final_winner:
                final_losers.append(semi_winner['winner'])
        
    final_winners.append(final_losers)
    sim_df.loc[sim_count] = final_winners
    sim_count += 1

sim_df
```
