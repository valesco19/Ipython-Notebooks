```python
def convert_int(x):
    try:
        return int(x)
    except:
        return np.nan

for col in score_df.columns:
    if col != 'player':
        score_df[col] = score_df[col].apply(lambda x: convert_int(x))
    
score_df = score_df.loc[score_df['year'] > 1990]
grouped_df = score_df[['player_id', 'year', 'total_score']].groupby(['year', 'player_id']).sum()
grouped_df.reset_index(inplace = True)
grouped_df
```

```python
years = grouped_df['year'].unique()
year_df = pd.DataFrame(columns = ['year', 'player_id', 'total_score', 'rank'])

for year in years:
    temp_df = grouped_df.loc[grouped_df['year'] == year]
    temp_df['rank'] = temp_df['total_score'].rank(axis = 0, method = 'min')
    temp_df = temp_df.sort('rank')
    year_df = year_df.append(temp_df)
    
year_df
```

```python
hole_ls = []
for col in score_df.columns:
    if 'hole' in col:
        hole_ls.append(col)
        
base_score_df = pd.DataFrame(columns = ['player_id', 'year', 'round', 'base_score'])
row_count = 0

for index, row in score_df.iterrows():
    total_score = 0
    streak_score = 0
    streak_total = 0
    birdie_streak = 0
    birdie_switch = 0
    bogie_streak = 0
    for h, hole_col in enumerate(hole_ls):
        score = row[hole_col]
        hole_num = int(hole_col.replace('hole_', ''))
        par = par_dict[hole_num]
        diff = score - par
        
        if diff == -3:
            score_name = 'double_eagle'
        elif diff == -2:
            score_name = 'eagle'
        elif diff == -1:
            score_name = 'birdie'
        elif diff == 0:
            score_name = 'par'
        elif diff == 1:
            score_name = 'bogey'
        elif diff >= 2:
            score_name = 'double_bogey'
        else:
            print('error', diff)
            
        if score_name == 'double_eagle' or score_name == 'eagle' or score_name == 'birdie':
            birdie_streak += 1
        else:
            birdie_streak = 0
            
        if birdie_streak > 2 and birdie_switch == 0:
            streak_total += dk_scoring['birdie_streak']
            birdie_switch = 1
            birdie_streak = 0
            
        if diff > 0:
            bogie_streak = 1
            
        if score == 1:
            streak_total += dk_scoring['hole_in_one']
        
        total_score += dk_scoring[score_name]
        
    if bogie_streak == 0:
        streak_total += dk_scoring['bogey_free_round']
    
    total_score += streak_total
                             
    base_score_df.loc[row_count] = [row['player_id'], row['year'], row['round_num'], total_score]
    row_count += 1                             
base_score_df.head()
```
