```python
import pandas as pd
import time
from datetime import datetime, timedelta
import feedparser
import os
import requests
import json
```

```python
os.chdir('D:/data/pypi_rss_feed')

feeds = {
    'pypi': 'https://pypi.python.org/pypi?%3Aaction=packages_rss'
    }

last_timestamp = datetime.now() - timedelta(days = 1)

while True:
    package_df = pd.DataFrame(columns = ['timestamp', 'summary', 'link'])
    count = 0
    
    for z in range(6):
        
        feed_json = []
        temp_df = pd.DataFrame(columns = ['timestamp', 'summary', 'link'])
        
        for feed, url in feeds.items():
            d = feedparser.parse(url)
            feed_json.append(d)

        for feed in feed_json:
            for entry in feed['entries']:
                timestamp = entry['published_parsed']
                timestamp = datetime.fromtimestamp(time.mktime(timestamp))
                link = entry['link']
                summary = entry['summary_detail']['value']
                
                temp_df.loc[count] = [timestamp, summary, link]
                
                count += 1
                
        temp_df = temp_df.loc[temp_df['timestamp'] > last_timestamp]
        temp_df.sort_values('timestamp', inplace = True)
        
        summary_values = temp_df['summary'].values
        link_values = temp_df['link'].values
        
        for s in range(len(summary_values)):
            print(summary_values[s], link_values[s])
            
        package_df = package_df.append(temp_df)
        package_df.sort_values('timestamp', inplace = True)
        package_df.reset_index(inplace = True, drop = True)
        package_df.drop_duplicates('summary', keep = 'first')
        last_timestamp = package_df['timestamp'].loc[len(package_df) - 1]
        
        time.sleep(7200)
    
    now = datetime.now()
    now_str = datetime.strftime(now, '%d_%m_%Y_%H_%M')
    filename = 'pypi_packages_' + now_str + '.csv'
    package_df.to_csv(filename, index = False) 
```
