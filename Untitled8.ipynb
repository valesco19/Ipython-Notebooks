```python
import os
import pandas as pd
import numpy as np
os.chdir('/home/valesco/Datasets/PGA_Data/pga_odds')
```

```python
with open('Bovada__2016_07_25_14_26.txt') as myfile:
    champ = myfile.readlines()
```

```python
new_list = []
for c in champ:
    new_list.append(c.replace('\n', ''))
```

```python
champ = new_list[:132]

name_ls = []
odds_ls = []
for i, c in enumerate(champ):
    if i%2 == 0:
        name_ls.append(c)
    else:
        odds_ls.append(c)
```

```python
champ_df = pd.DataFrame({'player': name_ls, 'odds': odds_ls})
```

```python
champ_df['odds'] = champ_df['odds'].map(lambda x: x.replace('+', ''))
champ_df
```
