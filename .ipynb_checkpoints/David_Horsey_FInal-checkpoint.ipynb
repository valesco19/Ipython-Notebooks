```python
def calc_miles(df):
    distance = vincenty(home_coords, (df['lat'], df['long'])).miles
    return distance

#calc distance traveled from hometown to tournament
geolocator = nm()
home_coords = geolocator.geocode('Alderley Edge, England', timeout = 10)
home_coords = home_coords[1]

course_loc['miles_dist'] = course_loc.apply(calc_miles, axis = 1)
course_loc
```

```python
# fuction to turn true false into numerical
def true_false_dummy(x):
    if x == 1:
        return 1
    else:
        return 0

# apply penalty stroke to the shot that the penalty occurred 
def apply_penalty(df):
    round_num = df['round_no']
    round_date = df['round_date']
    hole_num = df['hole_no']
       
    #Check to see if there is a next shot
    next_shot = df['shot_no'] + 1
    if len(horsey_df.loc[(horsey_df['round_no'] == round_num) & (horsey_df['round_date'] == round_date) &
                 (horsey_df['hole_no'] == hole_num) & (horsey_df['shot_no'] == next_shot)]) > 0:
        return 0
    else:
        #if there isn't a next shot, but there is a following shot, the previous shot was a penalty stroke
        third_shot = df['shot_no'] + 2
        if len(horsey_df.loc[(horsey_df['round_no'] == round_num) & (horsey_df['round_date'] == round_date) &
                     (horsey_df['hole_no'] == hole_num) & (horsey_df['shot_no'] == third_shot)]) > 0:
            return 1
        else:
            return 0
    
dummy_cols = ['putt', 'completed']

for col in dummy_cols:
    horsey_df[col] = horsey_df[col].apply(lambda x: true_false_dummy(x))
    
horsey_df['penalty'] = horsey_df.apply(apply_penalty, axis = 1)
horsey_df.replace({'Greenside Bunker': 'Sand', 'Green Fringe': 'Fringe', 
                  'Intermediate Rough': 'Intermediate','Fairway Bunker': 'Sand',
                  'Other': 'Rough'}, inplace = True)
    
horsey_df.head()
```

```python
## Function to round distances
def myround(x, base):
    return (int(base * round(float(x) / base)))

## Function to calculate strokes gained
def calc_sg(df):
    if df['completed'] == 0:
        return 0, np.nan, np.nan
    else:
        start_loc = df['lie_before']
        end_loc = df['lie_after']
        start_shot_dist = df['left_to_pin_before']
        end_shot_dist = df['left_to_pin_after']

        penalty = 0
        
        #Check if penalty stroke is applicable
        if df['penalty'] == 1:
            penalty = 1
            if end_loc == 'Out of Bounds':
                end_shot_dist = start_shot_dist
                end_loc = start_loc
                
            else:
                hole = df['hole_no']
                shot = df['shot_no'] + 2
                round_num = df['round_no']
                course_id = df['course_id']

                end_shot_dist = horsey_df['left_to_pin_before'].loc[(horsey_df['shot_no'] == shot) & 
                                    (horsey_df['round_no'] == round_num) & (horsey_df['course_id'] == course_id) &
                                    (horsey_df['hole_no'] == hole)].values[0]


                end_loc = horsey_df['lie_before'].loc[(horsey_df['course_id'] == course_id) & (horsey_df['shot_no'] == shot) &
                                    (horsey_df['round_no'] == round_num) & (horsey_df['hole_no'] == hole)].values[0]

        if start_loc == 'Green' or start_loc == 'Fringe':
            start_dist_rounded = myround(start_shot_dist / 12, 1)
            start_dist_col = 'PuttDistinfeet'
            
            if start_dist_rounded > 112:
                start_dist_rounded = 112
            if start_dist_rounded < 1:
                start_dist_rounded = 1
                
        elif start_loc in ['Tee', 'Fairway', 'Intermediate', 'Rough', 'Sand']:
            start_dist_rounded = myround(start_shot_dist / 36, 20)
            start_dist_col = 'Distance'
            
            if start_dist_rounded < 20:
                start_dist_rounded = 20
        else:
            start_dist_rounded = 0

        if end_loc == 'Green' or end_loc == 'Fringe':
            end_dist_rounded = myround(end_shot_dist / 12, 1)
            end_dist_col = 'PuttDistinfeet'
            
            if end_dist_rounded > 112:
                end_dist_rounded = 112
                
            if end_dist_rounded < 1 and end_shot_dist > 0:
                end_dist_rounded = 1
            
        elif end_loc in ['Tee', 'Fairway', 'Intermediate', 'Rough', 'Sand']:
            try:
                end_dist_rounded = myround(end_shot_dist / 36, 20) 
                end_dist_col = 'Distance'
                
                if end_dist_rounded < 20:
                    end_dist_rounded = 20
            except:
                return 0, np.nan, np.nan
        else:
            end_dist_rounded = 0
        
        if start_dist_rounded <= 2 and start_loc == 'Green':
            end_dist_rounded = 0

        if start_dist_rounded == 0:
            return 0, np.nan, np.nan
        else:
            start_sg = strokes_gained[start_loc].loc[strokes_gained[start_dist_col] == start_dist_rounded].values[0]

        if end_dist_rounded == 0:
            end_sg = 0
        else:
            end_sg = strokes_gained[end_loc].loc[strokes_gained[end_dist_col] == end_dist_rounded].values[0]
        
        return (start_sg - end_sg - 1 - penalty), start_loc, start_dist_rounded

    
horsey_df['stroke_gained'] = horsey_df.apply(calc_sg, axis = 1)
horsey_df[['stroke_gained', 'loc_bin', 'dist_bin']] = horsey_df['stroke_gained'].apply(pd.Series)
```

```python
## Create weather dictionary for the various tournaments, then since tee time date wasn't available, 

api = 'c7c2d4dea244baeca35b91ed97364062'

tourny_date_ls = horsey_df['round_date'].unique()

weather_dict = {}
for tourny_date in tourny_date_ls:
    #retrieve details for api call
    course_id = horsey_df['course_id'].loc[horsey_df['round_date'] == tourny_date].iloc[0]
    lat = course_loc['lat'].loc[course_loc['course_id'] == course_id].values[0]
    lon = course_loc['long'].loc[course_loc['course_id'] == course_id].values[0]
    date_object = datetime.datetime.strptime(tourny_date, '%Y-%m-%d')
    weather = forecastio.load_forecast(api, lat, lon, time = date_object, units = 'us')
    hourly = weather.hourly()

    #parse api response
    temp_weather_ls = []
    for hour in hourly.data:
        h = hour.d
        h['time_str'] = datetime.datetime.fromtimestamp(int(h['time'])).strftime('%m/%d/%Y %H:%M')
        temp_weather_ls.append(h)
    
    dict_key = str(course_id) + '_' + tourny_date
    weather_dict[dict_key] = temp_weather_ls
    
    
weather_summary = pd.DataFrame(columns = ['course_id', 'round_date', 'temp_avg', 'wind_avg', 'humidity_avg',
                                         'summary'])

#Average out hourly weather date for a daily average
for k, key in enumerate(weather_dict.keys()):
    course_id = key.split('_')[0]
    round_date = key.split('_')[1]
    temp_total = 0
    wind_total = 0
    humid_total = 0
    summary_ls = []
    hour_count = 0
    for hour in weather_dict[key]:
        hour_split = hour['time_str'].split(' ')[0].split('/')
        hour_date = hour_split[2] + '-' + hour_split[0] + '-' + hour_split[1]
        hour_hour = int(hour['time_str'].split(' ')[1].split(':')[0])
        
        if round_date == hour_date and hour_hour > 7 and hour_hour < 18:
            try:
                temp_total += hour['temperature']
                wind_total += hour['windSpeed']
                humid_total += hour['humidity']
                summary_ls.append(hour['summary'])
                hour_count += 1
            except:
                pass
    temp_avg = temp_total / hour_count
    wind_avg = wind_total / hour_count
    humid_avg = humid_total / hour_count
    summary = stats.mode(summary_ls)[0][0]
    weather_summary.loc[k] = [course_id, round_date, temp_avg, wind_avg, humid_avg, summary]

weather_summary
```

```python
## Calculate course par and average course distance

# groupby course_id and hole num for stroke 1 to get hole par and average distance
hole_info_df = horsey_df[['course_id', 'hole_no', 'left_to_pin_before', 'par', 'shot_no']].loc[horsey_df['shot_no'] \
            == 1].groupby(['course_id', 'hole_no']).mean()

hole_info_df.reset_index(inplace = True)
hole_info_df['hole_dist_yards'] = hole_info_df['left_to_pin_before'] / 36
hole_info_df = hole_info_df[['course_id', 'hole_no', 'par', 'hole_dist_yards']]
hole_info_df
```

```python
## Further group hole_info_df into course totals and join it to the course_loc df

course_totals_df = hole_info_df[['course_id', 'par', 'hole_dist_yards']].groupby(['course_id']).sum()
course_totals_df.reset_index(inplace = True)
course_loc = course_loc.merge(course_totals_df, on = 'course_id')
course_loc
```

```python
## Aggregate SG Totals grouped by starting location
agg_dict = {'stroke_gained': 'sum', 'round_date': 'count'}

horsey_sg_total = horsey_df[['course_id', 'round_no', 'round_date', 'stroke_gained', 
                             'loc_bin']].groupby(['course_id', 'round_no', 'loc_bin']).agg(agg_dict)

horsey_sg_total = horsey_sg_total.rename(columns = {'round_date': 'shot_count'})
horsey_sg_total = horsey_sg_total.unstack('loc_bin').reset_index()
horsey_sg_total.columns = horsey_sg_total.columns.droplevel()

horsey_cols = ['course_id', 'round_no', 'fairway_sg', 'fringe_sg', 'green_sg', 'intermediate_sg', 'rough_sg',
              'sand_sg', 'tee_sg', 'fairway_count', 'fringe_count', 'green_count', 'intermediate_count', 
               'rough_count', 'sand_count', 'tee_count']
horsey_sg_total.columns = horsey_cols
horsey_sg_total.fillna(0, inplace = True)
horsey_sg_total['total_sg'] = horsey_sg_total['fairway_sg'] + horsey_sg_total['fringe_sg'] + \
                 horsey_sg_total['green_sg'] + horsey_sg_total['intermediate_sg'] + horsey_sg_total['rough_sg'] + \
                 horsey_sg_total['sand_sg'] + horsey_sg_total['tee_sg']
        
horsey_sg_total['total_count'] = horsey_sg_total['fairway_count'] + horsey_sg_total['fringe_count'] + \
                 horsey_sg_total['green_count'] + horsey_sg_total['intermediate_count'] + horsey_sg_total['rough_count'] + \
                 horsey_sg_total['sand_count'] + horsey_sg_total['tee_count']

horsey_sg_total['rough_sg'] = horsey_sg_total['rough_sg'] + horsey_sg_total['intermediate_sg']
horsey_sg_total['rough_count'] = horsey_sg_total['rough_count'] + horsey_sg_total['intermediate_count']
horsey_sg_total['total_sg_shot'] = horsey_sg_total['total_sg'] / horsey_sg_total['total_count']
horsey_sg_total['fairway_sg_shot'] = horsey_sg_total['fairway_sg'] / horsey_sg_total['fairway_count']
horsey_sg_total['fringe_sg_shot'] = horsey_sg_total['fringe_sg'] / horsey_sg_total['fringe_count']
horsey_sg_total['green_sg_shot'] = horsey_sg_total['green_sg'] / horsey_sg_total['green_count']
horsey_sg_total['rough_sg_shot'] = horsey_sg_total['rough_sg'] / horsey_sg_total['rough_count']
horsey_sg_total['sand_sg_shot'] = horsey_sg_total['sand_sg'] / horsey_sg_total['sand_count']
horsey_sg_total['tee_sg_shot'] = horsey_sg_total['tee_sg'] / horsey_sg_total['tee_count']
        
horsey_cols = ['course_id', 'round_no', 'total_sg_shot', 'fairway_sg_shot', 'fringe_sg_shot', 'green_sg_shot',
              'rough_sg_shot', 'sand_sg_shot', 'tee_sg_shot', 'total_sg', 'fairway_sg', 'fringe_sg', 'green_sg', 'rough_sg',
              'sand_sg', 'tee_sg', 'total_count', 'fairway_count', 'fringe_count', 'green_count', 
               'rough_count', 'sand_count', 'tee_count']


horsey_sg_total = horsey_sg_total[horsey_cols]
horsey_sg_total.fillna(0, inplace = True)
horsey_sg_total
```

```python
horsey_sg_avg = pd.DataFrame(horsey_sg_total.mean())

total_cols = ['total_sg', 'fairway_sg', 'fringe_sg', 'green_sg', 'rough_sg', 'sand_sg', 'tee_sg']
shot_cols = ['total_sg_shot', 'fairway_sg_shot', 'fringe_sg_shot', 'green_sg_shot', 'rough_sg_shot',
             'sand_sg_shot', 'tee_sg_shot']
count_cols = ['total_count', 'fairway_count', 'fringe_count', 'green_count', 'rough_count', 'sand_count', 
              'tee_count']
shot_sg_avg = horsey_sg_avg.transpose()[shot_cols]
shot_sg_avg.columns = ['Total SG/Shot', 'Fairway SG/Shot', 'Fringe SG/Shot', 'Green SG/Shot', 
                       'Rough SG/Shot', 'Sand SG/Shot', 'Tee SG/Shot']
shot_sg_avg = shot_sg_avg.transpose().reset_index()
shot_sg_avg.columns = ['field', 'value']

total_sg_avg = horsey_sg_avg.transpose()[total_cols]
total_sg_avg.columns = ['Total SG', 'Fairway SG', 'Fringe SG', 'Green SG', 'Rough SG', 'Sand SG', 'Tee SG']
total_sg_avg = total_sg_avg.transpose().reset_index()
total_sg_avg.columns = ['field', 'value']

count_sg_avg = horsey_sg_avg.transpose()[count_cols]
count_sg_avg.columns = ['Total Shot Count', 'Fairway Shot Count', 'Fringe Shot Count', 'Green Shot Count', 
                        'Rough Shot Count', 'Sand Shot Count', 'Tee Shot Count']
count_sg_avg = count_sg_avg.transpose().reset_index()
count_sg_avg.columns = ['field', 'value']
```

```python
match_date_df = horsey_df[['course_id', 'round_no', 'round_date']].groupby(['round_date']).mean().reset_index()

horsey_final_round_df = pd.merge(horsey_sg_total, match_date_df[['round_date', 'course_id', 'round_no']],
                        left_on = ['course_id', 'round_no'],right_on = ['course_id', 'round_no'], how = 'inner')

horsey_final_round_df = horsey_final_round_df.merge(weather_summary, on = 'round_date')
horsey_final_round_df = horsey_final_round_df.rename(columns = {'course_id_x': 'course_id'})
horsey_final_round_df = horsey_final_round_df.merge(course_loc[['course_id', 'miles_dist', 'par', 
                            'hole_dist_yards']], on = 'course_id')
summary_dummies = pd.get_dummies(horsey_final_round_df['summary'])

horsey_final_round_df = pd.concat([horsey_final_round_df, summary_dummies], axis = 1)

horsey_final_round_df = horsey_final_round_df.drop(['summary', 'course_id_y'], axis = 1)
horsey_final_round_df
```

```python
def day_to_timestamp(x):
    stamp = time.mktime(datetime.datetime.strptime(x, '%Y-%m-%d').timetuple())
    return stamp

horsey_final_round_df['timestamp'] = horsey_final_round_df['round_date'].apply(lambda x: day_to_timestamp(x))


scale = StandardScaler()
scaled_round_df = horsey_final_round_df[['course_id', 'round_no']]
col_list = horsey_final_round_df.columns

for col in col_list[2:]:
    if col != 'round_date':
        scaled_round_df[col] = scale.fit_transform(horsey_final_round_df[col].values.reshape(-1,1))
```

```python
corr_df = pd.DataFrame(columns = ['Initial Col', 'Second Col', 'R value', 'P value'])

scaled_cols = ['round_no','temp_avg','wind_avg','humidity_avg','miles_dist','par','hole_dist_yards','Clear','Light Rain',
'Mostly Cloudy','Overcast','Partly Cloudy','timestamp']
cols_to_compare = ['total_sg_shot', 'fairway_sg_shot', 'rough_sg_shot', 'green_sg_shot', 'sand_sg_shot']

count = 0
for scaled_col in scaled_cols:
    for compare_col in cols_to_compare:
        slope, intercept, r_value, p_value, std_err = stats.linregress(scaled_round_df[scaled_col],
                                            scaled_round_df[compare_col])

        corr_df.loc[count] = [compare_col, scaled_col, r_value, p_value]
        count += 1

corr_df.sort_values(by = 'P value', inplace = True)
corr_df = corr_df.loc[corr_df['P value'] < 0.05]
corr_df.replace({'total_sg_shot': 'Total SG Per Shot', 'fairway_sg_shot': 'Fairway SG Per Shot', 
                'sand_sg_shot': 'Sand SG Per Shot', 'green_sg_shot': 'Green SG Per Shot', 
                'wind_avg': 'Wind Avg', 'round_no': 'Round Num'}, inplace = True)

corr_df = corr_df.round(4)
table = ff.create_table(corr_df)
py.iplot(table)
```

```python
agg_dict = {'stroke_gained': 'sum', 'loc_bin': 'count'}
horsey_putt = horsey_df[['stroke_gained', 'loc_bin', 'dist_bin']].loc[(horsey_df['loc_bin'] == 'Green')].groupby(['dist_bin']).agg(agg_dict)
horsey_putt.reset_index(inplace = True)
horsey_putt = horsey_putt.rename(columns = {'loc_bin': 'shots'})
horsey_putt['sg_per_putt'] = horsey_putt['stroke_gained'] / horsey_putt['shots']

horsey_putt = horsey_putt.merge(strokes_gained[['PuttDistinfeet', 'Green']], right_on = 'PuttDistinfeet', 
                    left_on = 'dist_bin', how = 'inner')

horsey_putt['horsey_difference'] = horsey_putt['Green'] - horsey_putt['sg_per_putt']
horsey_putt
```

```python
agg_dict = {'stroke_gained': 'sum', 'loc_bin':'count'}

horsey_long_game = horsey_df.loc[(horsey_df['loc_bin'] != 'Green')]
horsey_long_game = horsey_long_game.loc[horsey_long_game['loc_bin'] != 'Fringe']
horsey_dist_grouped = horsey_long_game[['stroke_gained', 'dist_bin', 'loc_bin']].groupby(
                        ['dist_bin']).agg(agg_dict)

horsey_dist_grouped.reset_index(inplace = True)
horsey_dist_grouped = horsey_dist_grouped.rename(columns = {'loc_bin': 'shots'})
horsey_dist_grouped['sg_per_shot'] = horsey_dist_grouped['stroke_gained'] / horsey_dist_grouped['shots']

horsey_dist_grouped
```

```python
strokes_gained['stroke_combined'] = (strokes_gained['Fairway'] + strokes_gained['Intermediate'] + 
                                strokes_gained['Rough'] + strokes_gained['Sand']) / 4

horsey_dist_grouped = horsey_dist_grouped.merge(strokes_gained[['Distance', 'stroke_combined']], left_on = 'dist_bin', 
                         right_on = 'Distance', how = 'left')

horsey_dist_grouped['horsey_difference'] = horsey_dist_grouped['stroke_combined'] - horsey_dist_grouped['sg_per_shot']
horsey_dist_grouped
```

```python
putt_dist = list(horsey_putt['PuttDistinfeet'].iloc[::-1])
avg_strokes = list(horsey_putt['Green'].iloc[::-1])

#smooth out Horsey's putt due to sample size issues

horsey_putts = list(horsey_putt['horsey_difference'].rolling(window = 6).mean().iloc[::-1])




trace0 = go.Scatter(x = putt_dist, y = avg_strokes, name = 'Tour Average', 
                    line = dict(color = ('rgb(205, 12, 24)'),
                                width = 2.5))

trace1 = go.Scatter(x = putt_dist, y = horsey_putts, name = 'David Horsey', 
                   line = dict(color = ('rgb(22, 96, 167)'),
                    width = 2.5))

data = [trace0, trace1]

layout = dict(title = 'David Horsey vs Tour Average Strokes Gained Putting', 
            xaxis = dict(title = 'Distance in feet'),
            yaxis = dict(title = 'Average Putts to Complete the Hole'))

figure = dict(data = data, layout = layout)
py.iplot(figure, filename = 'horsey_sg_putt')
```
