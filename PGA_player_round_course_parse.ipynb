```python
os.chdir('/home/valesco/Datasets/PGA_Data/PGA/')

player_course_df = pd.DataFrame(columns = ['player_id', 'player_name', 'tourny_id', 'tourny_year',
                                           'round_id', 'round', 'course_num', 'course_name'])
count = 0

tournament_ls = glob('*')

for tourny in tqdm(tournament_ls):
    try:
        os.chdir('/home/valesco/Datasets/PGA_Data/PGA/' + tourny)
        tee_json = open(tourny + '_teetimes.json')
        tee_data = json.load(tee_json)

        tourny_prefix = tourny.split('_')[0]
        tourny_year = tourny.split('_')[1]
        for r, rnd in enumerate(tee_data['tournament']['rounds']):
            round_str = str(r + 1)
            for course in rnd['courses']:
                course_num = str(course['CourseNumber'])
                course_name = course['CourseName']
                for segment in course['segments']:
                    for group in segment['groups']:
                        for player in group['players']:
                            player_id = player['TournamentPlayerID']
                            player_name = player['LFName']

                            round_id = '{}_{}_{}_{}'.format(player_id, tourny_year, tourny_prefix, round_str)

                            player_course_df.loc[count] = [player_id, player_name, tourny_prefix, tourny_year,
                                                        round_id, round_str, course_num, course_name]
                            count += 1
    except Exception as e:
        pass
                    

os.chdir('/home/valesco/Datasets/PGA_Data/pga_master_files')
player_course_df.to_csv('player_round_course.csv', sep = ';', index = False)
```
