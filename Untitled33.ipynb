```python
def myround(x, base):
    return (int(base * round(float(x) / base)))

sal_df['dist_bin'] = sal_df['shot_dist'].apply(lambda x: myround(x / 36, 20))
grouped = sal_df[['tournament_id', 'year', 'shot_loc', 'dist_bin', 'stroke_gained']].groupby(['tournament_id', 'year',
                                    'shot_loc', 'dist_bin']).agg({'stroke_gained':  'count'})
grouped.reset_index(inplace = True)
grouped
#grouped.to_csv('/home/valesco/Datasets/PGA_Data/pga_temp/st_jude.csv')
```

```python
count_rank_df = pd.DataFrame(columns = ['tournament_id', 'year', 'shot_loc', 'dist_bin', 'stroke_gained', 'rank'])

dist_range = [20, 40, 60, 80, 100, 120, 140, 160, 180, 200, 220, 240, 260, 280]
lie_ls = ['Rough', 'Fairway']

for lie in lie_ls:
    for dist in dist_range:
        temp_df = grouped.loc[(grouped['dist_bin'] == dist) & (grouped['shot_loc'] == lie)].sort_values(by = 'stroke_gained', ascending = False)
        temp_df['rank'] = temp_df['stroke_gained'].rank(ascending = False)
        temp_df = temp_df.loc[temp_df['tournament_id'] == 25]

        count_rank_df = pd.concat([count_rank_df, temp_df], axis = 0)
        
count_rank_df.sort_values('rank')
```

```python
def return_tid(x):
    tourny_ids = pd.read_csv('/home/valesco/Datasets/PGA_Data/pga_master_files/tournament_ids.csv', sep = ';')
    tourny_name_ls = tourny_ids['tournament_name'].unique()
    matched = process.extractOne(x, tourny_name_ls)
    
    if matched[1] > 59:
        temp_df = tourny_ids.loc[tourny_ids['tournament_name'] == matched[0]].sort_values('year', ascending = False)
                                                                                
        
        return temp_df['tournament_id'].values[0]
    
    else:
        print('{} was not matched'.format(x))
        return np.nan
    
tourny_unmatched_ls = sal_df['tournament_name'].unique()

matched_dict = {}
for unmatch in tourny_unmatched_ls:
    temp_tid = return_tid(unmatch)
    matched_dict[unmatch] = temp_tid
    
matched_dict
```

```python
matched_dict['British Open'] = 100

def return_tid(x):
    return matched_dict[x]

sal_df['tournament_id'] = sal_df['tournament_name'].apply(lambda x: return_tid(x))
sal_df
```

```python
agg_dict = {'actual_points': ['mean', 'std']}
grouped_df = sal_df[['tournament_id', 'year', 'actual_points']].groupby(['tournament_id', 'year']).agg(agg_dict)
grouped_df.reset_index(inplace = True)
grouped_df
```

```python
def return_tourny_avg(row):
    tid = row['tournament_id']
    year = row['year']
    
    avg_points = grouped_df['actual_points']['mean'].loc[(grouped_df['tournament_id'] == tid) & (grouped_df['year'] == year)].values[0]
    
    return avg_points

def return_tourny_std(row):
    tid = row['tournament_id']
    year = row['year']
    
    std_points = grouped_df['actual_points']['std'].loc[(grouped_df['tournament_id'] == tid) & (grouped_df['year'] == year)].values[0]
    
    return std_points

def norm_points(row):
    return (row['actual_points'] - row['tourny_avg_points']) / row['tourny_std_points']
    
sal_df['tourny_std_points'] = sal_df.apply(return_tourny_std, axis = 1)
sal_df['tourny_avg_points'] = sal_df.apply(return_tourny_avg, axis = 1)
sal_df['norm_points'] = sal_df.apply(norm_points, axis = 1)
sal_df
```

```python
pids_ls = sal_df['player_id'].unique()

for pid in tqdm(pids_ls):
    temp_df = sal_df.loc[sal_df['player_id'] == pid]
    try:
        correct_player = temp_df['player_name'].mode()[0]

        player_ls = temp_df['player_name'].unique()

        for player in player_ls:
            if player != correct_player:
                sal_df['player_id'].loc[sal_df['player_name'] == player] = np.nan
    except:
        pass
```
